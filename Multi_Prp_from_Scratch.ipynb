```python
from sklearn.datasets import load_iris
import numpy as np 
from sklearn.preprocessing import OneHotEncoder
```

```python
iris = load_iris()
X = iris.data.T
Y_i = iris.target.reshape(-1,1)
```

```python
encoder =  OneHotEncoder(sparse_output = False)
Y = encoder.fit_transform(Y_i).T
```

```python
X.shape
```

```python
limit_1 = np.sqrt(2/64)
W1 = np.random.normal(0, limit_1, (64,4))
B1 = np.zeros((64,1))
limit_2 = np.sqrt(2/32)
W2 = np.random.normal(0, limit_2, (32,64))
B2 = np.zeros((32,1))
limit_3 = np.sqrt(2/16)
W3 = np.random.normal(0, limit_3, (16,32))
B3 = np.zeros((16,1))
limit_4 = np.sqrt(6/16+3)
W4 = np.random.normal(-limit_4, limit_4, (3,16))
B4 = np.zeros((3,1))
```

```python
lr = 0.01
epochs = 100
nr_correct = 0 
batch_size = 64 
for epoch in range(epochs):
    permutation = np.random.permutation(X.shape[1])
    X_shuffled = X[:, permutation]
    Y_shuffled = Y[:, permutation]
    checkpoint = 0
    for checkpoint in range(0, X.shape[1], batch_size):
        x = X_shuffled[:, checkpoint:checkpoint + batch_size]
        y_true = Y_shuffled[:, checkpoint:checkpoint + batch_size]
        H_L_1 = np.dot(W1,x) + B1
        A_H_L_1 = np.maximum(0,H_L_1)
        H_L_2 = np.dot(W2,A_H_L_1) + B2
        A_H_L_2 = np.maximum(0,H_L_2)
        H_L_3 =  np.dot(W3,A_H_L_2) + B3
        A_H_L_3 = np.maximum(0,H_L_3)
        O_L = np.dot(W4, A_H_L_3) + B4
        exp_logits = np.exp(O_L - np.max(O_L, axis=0, keepdims=True))
        A_O_L = exp_logits / np.sum(exp_logits, axis=0, keepdims=True)
        predictions = np.argmax(A_O_L, axis=0)
        labels = np.argmax(y_true, axis=0)
        nr_correct += np.sum(predictions == labels)
        epsilon = 1e-8
        y_pred = np.clip(A_O_L, epsilon, 1 - epsilon)
        loss = -np.mean(np.sum(y_true * np.log(y_pred), axis=0))
        dZ4 = y_pred - y_true
        dW4 = np.dot(dZ4,A_H_L_3.T) / batch_size
        dB4 = np.sum(dZ4, axis=1, keepdims=True) / batch_size
        dA3 = np.dot(W4.T,dZ4)
        dZ3 = dA3*(H_L_3 > 0)
        dW3 = np.dot(dZ3,A_H_L_2.T)/batch_size
        dB3 = np.sum(dZ3, axis=1, keepdims=True)/ batch_size
        dA2 = np.dot(W3.T,dZ3)
        dZ2 = dA2*(H_L_2 > 0)
        dW2 = np.dot(dZ2,A_H_L_1.T)/batch_size
        dB2 = np.sum(dZ2, axis=1, keepdims=True)/ batch_size
        dA1 = np.dot(W2.T,dZ2)
        dZ1 = dA1*(H_L_1 > 0)
        dW1 = np.dot(dZ1,x.T)/batch_size
        dB1 = np.sum(dZ1, axis=1, keepdims=True)/ batch_size
        W4 += -lr*dW4
        B4 += -lr*dB4
        W3 += -lr*dW3
        B3 += -lr*dB3
        W2 += -lr*dW2
        B2 += -lr*dB2
        W1 += -lr*dW1
        B1 += -lr*dB1
        checkpoint += batch_size
        print(f"Epoch {epoch+1}/{epochs} - Accuracy: {round((nr_correct / X.shape[0]) * 100, 2)}%")
print(f"\nFinal Accuracy: {round((nr_correct / (epochs * X.shape[1])) * 100, 2)}%")
nr_correct = 0
```
